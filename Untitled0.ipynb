```python
df=pd.merge(marketing,ca,on='Date',how='outer')
print(df)
```

```python
#df.dtypes
print("Valeurs manquantes par colonne :")
print(df.isnull().sum())

for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna('Inconnu', inplace=True)
    elif df[col].dtype == 'float64':
        df[col].fillna(0, inplace=True)

print("DonnÃ©es aprÃ¨s nettoyage :")
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y',errors='coerce')
print(df)
print('Valeurs manquantes par colonne')
print(df.isnull().sum())
```

```python
df.rename(columns={'Day_Name': 'Day'}, inplace=True)
print(df)
```

```python
df['Mois'] = df['Date'].dt.month
print(df)
```

```python
visiteurs_par_mois = df.groupby('Mois')['Visitors'].sum()
print(visiteurs_par_mois)
mois_max = visiteurs_par_mois.idxmax()
print(f"Le mois avec le plus de visiteurs est le {mois_max}e de l'annÃ©e")
```

```python
avg_revenue = df['Revenue'].mean()
avg_marketing_spend = df['Marketing Spend'].mean()
avg_visitors = df['Visitors'].mean()
print(avg_revenue)
print(avg_marketing_spend)
print(avg_visitors)
```

```python
promo_averages = df.groupby('Promo')[['Revenue', 'Marketing Spend', 'Visitors']].mean()
print(promo_averages)
```
